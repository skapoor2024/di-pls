```python
def uniform_selection_2(dataframe, bin_width, target, min_samples,min_,max_):
    select = []
    bins = int((max_ - min_)/ bin_width)
    for i in range(bins + 1):
        condition = (dataframe[target] >= min_ + i * bin_width) & (dataframe[target] 
                                                            < min_ + (i + 1) * bin_width)
        if dataframe[condition].shape[0] >= min_samples:
            filtered = dataframe[condition].sample(min_samples)
            select.append(filtered)
        else:
            select.append(dataframe[condition])
    final = pd.concat(select)
    return final
```
